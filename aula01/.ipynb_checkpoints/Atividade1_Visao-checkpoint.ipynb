# Equalização de histograma

```python
rintin = cv2.imread("RinTinTin.jpg")
```

```python
rintin.shape
```

```python
rintin_gray = cv2.cvtColor(rintin, cv2.COLOR_RGB2GRAY)
```

```python
rintin_gray.shape
```

```python
plt.imshow(rintin_gray, cmap="Greys_r", vmin=0, vmax=255)
```

```python

hist,bins = np.histogram(rintin_gray.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
img = rintin_gray
saida = np.zeros((240, 320), dtype=np.uint8)


for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        saida[i][j] = 4.39*img[i][j]

plt.imshow(saida, cmap = "Greys_r", vmin = 0, vmax = 255)
```

```python
plt.hist(saida.flatten(),256,[0,256], color = 'y')
plt.xlim([0,256])
plt.show()
```

```python
saida_nova = saida.copy()

for i in range(saida.shape[0]):
    for j in range(saida.shape[1]):
        saida_nova[i][j] = saida[i][j] - 30
        
plt.imshow(saida_nova, cmap = 'Greys_r', vmin = 0, vmax = 255)
```

# Atividade 3


Abra a imagem `hall_box_battery_atividade3.png`  e a converta para tons de cinza de `0` a `255`.

Forneça como saída uma uma cópia da imagem  em que cada pixel  recebe o módulo da subtração entre o pixel que vem depois e o que vem antes dele (na horizontal).

Obs.: O conceito de *antes* e *depois* pode ser na horizontal ou na vertical

Apresente a imagem resultado, deve ficar parecida com a imagem `gabarito_atividade_3.png`.

<img src="gabarito_atividade_3.png" width=400>

```python
hallbox = cv2.imread("hall_box_battery_atividade3.png")
plt.imshow(hallbox)
```

```python
hallbox_gray = cv2.cvtColor(hallbox, cv2.COLOR_BGR2GRAY)
plt.imshow(hallbox_gray, cmap = "Greys_r", vmin = 0, vmax = 255)
```

```python
saida = hallbox_gray.copy()
saida.shape
```

```python
hallbox_gray[5][6] - hallbox_gray[5][5]
```

```python
saida = np.zeros((576, 1024), dtype=np.uint8)

for i in range(hallbox_gray.shape[0]):
    for j in range(1, hallbox_gray.shape[1] - 1):
        subtracao = abs(int(hallbox_gray[i][j+1]) - int(hallbox_gray[i][j-1]))
        if subtracao < 0:
            saida[i][j] = subtracao*(-1)
        else:
            saida[i][j] = subtracao

plt.imshow(saida, cmap ='Greys_r', vmin = 0, vmax = 255)
```
